# Analysis of a mismatched LNA

In this notebook, we will discuss a mismatched LNA and analyze the limitations of different measurements of its noise performance.

Recall that in the last section in [Y_factor.ipynb](./Y_factor.ipynb), we performed a Y-factor measurement to derive the gain and the noise temperature of the LNA. 

The incoming voltage wave from the source is partially reflected at the input of the LNA due to impedance mismatch, and the resistor power being delivered to the LNA is,
$$
P_{net} = P_{inc} - P_{ref} \\
P_{net} \propto |V^+|^2 - |V^-|^2 = (1 - |\Gamma|^2)|V^+|^2
$$
where the incident wave amplitude $V^+$ is $kTB$ (take this for granted now), where T is the physical temperature of the source.

Be careful that $P_{net}$ is _NOT_ proportional to total voltage amplitude,
$$
V_{tot} = V^+ + V^- = (1+\Gamma) V^+
$$

Note that $\Gamma$ is not necessarily the S11 of LNA. It is defined as, 
$$
\Gamma = \frac{Z_{LNA} - Z_{source}}{Z_{source} + Z_{LNA}}
$$

In most cases, the source impedance is 50 Ohm. However, during Y-Factor measurement, the source is a 50 Ohm resistor extended by a short coax cable (around 22 cm).

We can measure the impedance of $Z_{source}$ and see how it changes $(1-|\Gamma|^2)$ compared to the ideal case of $Z_{source} = 50 \Omega$.
